```python
import numpy as np
import matplotlib.pyplot as plt

# Load numbers from a text file (one column)
data = np.loadtxt("/home/dave/CHT-model/Optimized_LBE_petsc_pressure/error_BGK.txt")   # change file name if needed

# x-axis: indices from 0 to len(data)-1
x = np.arange(len(data))

# Plot as a thin line (no markers)
plt.figure(figsize=(8,4))
plt.plot(x, data, linewidth=0.8, color="blue")   # thin line
plt.xlabel("Index")
plt.ylabel("Value")
plt.title("Residual BGK LBE, Re = 1000, Poiseuille flow")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

```
